# Cuerda Vibrante de Extremos Fijos

### Visualización de los Modos Normales.


Prof. Mario I. Caicedo

* El estudio d ela dinámica de la cuerda cargada nos enseñó que un tal sistema con $N$ masas, posee exaxtamnte $N$ modos normales de oscilación

* El sistema permite hacer un paso al límite del contínuo

$$N\rightarrow\infty$$

las coordenadas de las masas ($\psi_n(t)$) se convierten en una función de dos variables $u(x,t)$ que representa las oscilaciones transversales de una cuerda vibrante.

* En el problema de la cuerda vibrante cn extremos fijos, aparece una cadena infinita de modos normales.



* Si nos limitamos al caso en que una de las condiciones iniciales es
$$\frac{\partial{}u(x,t=0)}{\partial{}t}=0$$

Los modos normales adoptan la forma:

$$u_p(x,t)=sen(k_0x)cos(p\omega_0t)\,\quad{}p=1,2,3,\dots$$

$$k_0=\pi/L\,,\quad{}\omega_0=vk_0$$

NOTA: El p-ésimo modo posee $p-1$ puntos que permanecen inmóviles (nodos). 

**Ejercicio**

Utilice este cuaderno como base para visualizar los modos de la cuerda con extremos libres

Bibliotecas

------------------------------------------------------

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Definición de los modos

```python
L=2.
v=.04
```

```python
p=float(input('Introduzca un antero para escoger el modo que quiere visualizar '))
```

```python
def u(x,n):
    
    k_0=np.pi/L
    omega_0=v*k_0
    
    u = np.sin(p*k_0*x)*np.cos(p*omega_0*n)
    
    return u
```

* Creando la figura y los ejes

```python
fig = plt.figure(figsize=(12,5))
ax1 = plt.subplot(1,1,1)   

ax1.set_xlim(( 0, 2))            
ax1.set_ylim((-2, 2))
ax1.set_xlabel('x')
ax1.set_ylabel('u(x,t)')

txt_title = ax1.set_title('')
line1, = ax1.plot([], [], 'b', lw=2)

ax1.legend(['u(x,t)']);
```

* Animación

```python
def init():
    line1.set_data([], [])
    return line1,
```

```python
def drawframe(n):
    x = np.linspace(0, 2, 1000)
    
    y1 = u(x,n)
    
    line1.set_data(x, y1)
    txt_title.set_text('Frame = {0:4d}'.format(n))
    return line1,
    
```

```python
from matplotlib import animation

# blit=True re-draws only the parts that have changed.
#anim = animation.FuncAnimation(fig, drawframe, frames=100, interval=20, blit=True)
anim = animation.FuncAnimation(fig, drawframe, frames=100, interval=20, blit=True)
```
